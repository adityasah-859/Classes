# Revisit List

### Various ways to create list

```python
list1 = []
print(list1)
print(type(list1))
```

```python
list2 = list(range(0,10,2))
print(list2)
print(type(list2))
```

```python
str1 = "Learning Python is Fun"
l = str1.split()
print(l)
print(type(l))
```

```python
list3 = []
print(list3, len(list3))

list3.append(1)
list3.append(20.5)
list3.append("Hello")
list3.append(True)
print(list3, type(list3))
```

```python
l4 = []
for i in range(1,1000):
    if i % 100 == 0:
        l4.append(i)
print(l4)
```

### Slicing in list

```python
print(list3[2:4])
```

```python
print(list3[1:4:2])
```

### Reverse list

```python
print(list3[::-1])
```

### Concatenation of list


```python
l4 = list(range(10,70,5))
str1 = "Learning python is fun always"
l5 = str1.split()
l6 = l4 + l5
print(l6)
```

### Copying a List

#### Reference

```python
l5 = [5, 6, 7]
l6 = l5 #<--- Dono same jagah pe point krta h aur ek jagah change krne pe dono change hoga
print(l5)
print(l6)
print(l5 is l6)
```

```python
l5.append(232)
print(l5)
print(l6)
```

#### Shallow Copy

```python
l7 = [5,6,10]
l8 = l7.copy() #<-- ek jagah change krne se dusre ko fark nhi padta kyunki dono alag alag jagah point kr rha h
print(l7)
print(l8)
print(l7 is l8)
```

```python
l7.append(123)
print(l7)
print(l8)
```

```python
l9 = [8,9,0]
l10 = [8,9,0]
print(l9 is l10)
l9.insert(3,100)
print(l9)
```

### List Functions

```python
l = [1,4,0,99,27,"Hello","World","Aditya"]
sorted_l = sorted(l,key=str)
print(sorted_l)
```

```python
list1 = ['a', 'b', 'c', 'd', 'e', 'f', 'a', 'a']

# Length
print(len(list1))

# Count occurrences
print(list1.count('a'))
print(list1.count('b'))

# Append
list1.append('Hello')
print(list1)

# Insert at index
list1.insert(1, 8888)
print(list1)

# Remove first occurrence
list1.remove('a')
print(list1)
     
```

## List Comprehension (Compact syntax for creating a list)


```python
s = [x * x for x in range(1,11)]
print(s)
```

### list = [expression for item in list if condition]

```python
s2 = [x for x in s if x % 2 == 0]
print(s2)
```

#### Working with words

```python

words = ['Identifiers', 'Keywords', 'Datatypes', 'Operators']
c = [w[0] for w in words]
print(c)
```

```python
str2 = "the quick brown fox jumps over the lazy dog"
words = str2.split()
a = [(w.upper(),len(w)) for w in words]
print(a)
```

```python
list1 = ['a', 'b', 'c', 'd', 'e', 'f']
list2 = ['a', 'b', 'c']
list3 = [e for e in list1 if e not in list2]
print(list3)
```

### Packing and unpacking

```python
a,b = [1,2]
print(a,b)
```

```python
[a,b,e,r] = [1,4,5,6] 
print(a,b,e,r)
```

# Revisiting Tuples

```python
t1 = 1,2,3,4,5
print(t1)
print(type(t1))
```

## Single element Tuples

```python
t2 = (10)
print(type(t2))

t3 = (10,)
print(type(t3))
print(t3)
```

## Converting list to tuples


```python
l = [10,9,24,2,23]
t4 = tuple(l)
print(t4)
print(type(t4))
```

## Tuple Operations

```python
t5 = t3 + t4
print(t5)
print(type(t5))
```

```python
t6 = t5*9
print(t6)
```

```python
print('Length of t6', len(t6))
print(t6.count(10))  # number of occurrences
print(t6.index(10))  # first occurrence

print(sorted(t6))  # sort elements
print(sorted(t6, reverse=True))  # reverse order

print(min(t6))
print(max(t6))
     
```

## Tuple Packing and Unpacking

```python
a = 10; b = 20; c = 30; d = 40
t8 = a,b,c,d, #<-- packing
print(t8)
print(type(t8))
```

```python
x,y,z,p = t8 #<--- Unpacking
print(x,y,z,p)
```

```python
st = "Adi"
a,b,c = st
print(a,b,c)
```

```python
a,*b = [10,23,5,12,65,90]
print(a)
print(b)
print(type(b))
```

```python
a,*b = (20,23,5,12,65,90)
print(a)
print(b)
print(type(b))
```

```python
*a,b = (20,23,5,12,65,90)
print(a)
print(b)
print(type(b))
```

# Revisiting Dictionary


```python
d1 = {}
print(d1)
print(type(d1))
```

### Adding and updating

```python
d2 = dict()
d2['Stock'] = 8
d2['Savings'] = 5
print(d2)
```

```python
d2['Stock'] = 9
print(d2)
```

### Shallow Copy

```python
d3 = d2.copy()
print(d3)
print(type(d3))
```

### deep copy


```python
d4 = d3
print(d4 is d3)
```

### Deleting

```python
del d4
```

### Clearing

```python
d3.clear()
print(d3)
```

### Keys and Values

```python
print(d2)
print(d2.keys())
print(d2.values())
```

```python
for k, v in d2.items():
    print(k, "-", v)
```

# Revisiting Sets

### Creating a set

```python
s = {}
print(s)
print(type(s))
```

```python
se2 = set()
print(type(se2))
```

```python
se3 = set(range(1,10))
print(se3)
print(type(se3))
```

### Add a element

```python
se3.add(99)
print(se3)
```

### Update 

```python
l1 = ['a', 'b']
l2 = ['x', 'y']
se3.update(l1,l2)
print(se3)
```

### Shallow copy

```python
se4 = se3.copy()
print(se4)
```

### remove pop clear

```python
se4.remove(99)
print(se4)
print(se3)
```

```python
se4.pop() #<--- Remove a random element from set
print(se4)
```

```python
se4.clear() #<--- Removes all element from set
print(se4)
```
