## Iterate through nested dictionaries

### Give name of the store1 and all the items in it if the item is a laptop then give its price

```python
print(market['store1']['name'])
items = market['store1']['items']
for e in items:
    if e['name'] == 'Laptop':
        print(e['name'],"-", e['price'])
    else:
        print(e['name'])
```

## Shallow copy (outer object is new, inner mutable objects are shared)

### Flat List

```python
l1 = [1,2,3]
l2 = l1.copy()
print('Before changes')
print('l1 :',l1)
print('l2 :',l2)
print('After changes with appending 4 in l2')
l2.append(4)
print(l1)
print(l2)
```

### Nested List

```python
l1 = [1,2,[3,4]]
l2 = l1.copy() #<-- Nested list mein jo andr ka list hota h woh reference hota h not shallow copy

print('Before changes')
print(l1)
print(l2)

l1[2].append(99)

print('\n After l1[2].append(99)')
print('l1 :', l1)
print('l2 :',l2)

l1.append(4)
print('\n After l1.append(4)')
print('l1 :',l1)
print('l2 :',l2)
```

### Deep copy(both outer and inner objects are independent)

```python
import copy
l1 = [1,2,[3,4]]
l3 = copy.deepcopy(l1) #<--- here the outer and inner both are individual

print('Before changes')
print(l1)
print(l3)

l1[2].append(10222)

print('\nafter l1[2].append(10222)')
print('l1: ',l1)
print('l2: ',l2)
```

### Upacking tuple since function always returns a tuple

```python
def calc(num1,num2):
    sum = num1+ num2
    prod = num1* num2
    div = num1/num2
    mod = num1%num2
    return sum,prod,div,mod

a,b,c,d = calc(2,4)
print(2,4)
print("sum: ",a)
print("product: ",b)
print("divide: ",c)
print("Modulo: ",d)

print(type(calc(2,4)))
print(id(calc(2,4)))
```

### Factorial of a number

#### without recurion

```python
def fact(num):
    result = 1
    while num >= 1:
        result*=num
        num-=1
    return result

print(fact(5))
```

#### With recursion

```python
def fact(num):
    if num == 0:
        result = 1
    else: 
        result =  num*fact(num-1)
    return result
print(fact(5))
```

# Scope of a variable

## Global variable

```python
a = 10

def f1():
    print(a)

def f2():
    print(a)

f1()
f2()
print(a)
```

## Local Variable


```python
def f1():
    ax = 21
    print(ax)
    
def f2():
    print(ax)

f1()
f2()
```

```python
def f1():
    global asd 
    asd = 213
    print(asd)

def f2():
    print(asd)

f1()
f2()
```

### Filter(using the function given it checks whether the element needs to be added or not)

```python
def isEven(num):
    if num % 2 == 0:
        return True
    return False

l1 = [1,23,45,6,3,6,4,3,26,89,76,90]
l2 = list(filter(isEven,l1))
print(l1)
print(l2)
```

```python
l3 = list(filter(lambda x: x % 2 ==0,l1))
print(l3)
```

### Map (same function is run through all the elements in the list)

```python
l1 = [1,2,3,4,5,6,7,8,5,34,65]
def double_(num):
    return num*2

l2 = list(map(double_,l1))
print(l2)
```

```python
l3 = list(map(lambda x : x*2,l1))
print(l3)
```

### Reduce (Cumulatively applies function in throughout the list)

```python
from functools import reduce
l7 = [1,34,5,12,76,98]
res = reduce(lambda x, y: x + y,l7)
print(res)
```

```python
from functools import reduce

l7 = [1, 2, 3, 4, 5]
res = reduce(lambda x, y: x + y, l7)
print(res)
```

# File Handling

## Writing a file

```python
f = open('testfile.txt', 'w')

print('File Name:', f.name)
print('File Mode:', f.mode)
print('Is the file readable:', f.readable())
print('Is the file writable:', f.writable())
print('Is the file closed:', f.closed)
f.close()
print('Is the file closed:', f.closed)
```

```python
f1 = open('testfile2.txt','w')
```

```python
for i in range(100):
    f1.write("This is a sample text" + str(i+1) + '\n')
f1.close()
print("Is the file closed",f1.closed)
```

### Context manager which itself closes the file without closing manually(Better Approach)

```python
with open("testfile.txt",'a') as f1:
    for i in range(12):
        f1.write('Phewwwwwwww'+str(i+1)+'\n')
print('Is the file closed', f1.closed)
```

## Reading a file

```python
f2 = open('testfile2.txt','r')
print("File name: ",f2.name)
print('File Mode:', f2.mode)
print('Is the file readable:', f2.readable())
print('Is the file writable:', f2.writable())
lines = f2.readlines()
for line in lines:
    print(line.strip())   # strip() removes newline characters
f2.close()
print('Is the file closed:', f2.closed)
```
