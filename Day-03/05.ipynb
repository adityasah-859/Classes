```python
import random
def get_stock_rate():
    return random.randint(-8,20)

def get_savings_rate():
    return random.randint(2,7)

def get_amount(principal,rate_fn,years):
    amount = principal
    for _ in range(years):
        amount+=amount*rate_fn()/100
    return amount
    

def run_simulation():
    years = 20
    investments = [('stocks',500),('savings',500)]
    rates = {'stocks':get_stock_rate,'savings':get_savings_rate}
    total = 0

    for item in investments:
        category = item[0]
        amount = item[1]
        rate_fn = rates[category]
        total += get_amount(amount,rate_fn,years)
    return total

outputs = []
for run in range(10000):
    outputs.append(run_simulation())


print("Max value of investment:",max(outputs))
print("Min value of investment:", min(outputs))
print("Average value of investment:",sum(outputs)/len(outputs))
```

## Iteration 4

### Write results to a file

```python
import datetime
from datetime import datetime

import random
def get_stock_rate():
    return random.randint(-8,20)

def get_savings_rate():
    return random.randint(2,7)

def get_amount(principal,rate_fn,years):
    amount = principal
    for _ in range(years):
        amount+=amount*rate_fn()/100
    return amount
    

def run_simulation():
    years = 20
    investments = [('stocks',500),('savings',500)]
    rates = {'stocks':get_stock_rate,'savings':get_savings_rate}
    total = 0

    for item in investments:
        category = item[0]
        amount = item[1]
        rate_fn = rates[category]
        total += get_amount(amount,rate_fn,years)
    return total

outputs = []
for run in range(10000):
    outputs.append(run_simulation())



with open('simulation_output.txt', 'a') as f:
    f.write(50 * '-' + '\n')
    f.write('Execution Time: ' + str(datetime.now()) + '\n')
    f.write(50 * '-' + '\n')
    f.write("Maximum value of investment: " + str(max(outputs)) + '\n')
    f.write("Minimum value of investment: " + str(min(outputs)) + '\n')
    f.write("Average value of investment: " + str(sum(outputs) / len(outputs)) + '\n') 
    f.write('\n')   
```
